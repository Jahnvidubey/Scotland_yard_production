```python
import pandas as pd

# Read the Excel file into a DataFrame
data = pd.read_excel("Scotland_Yard_Productions.xlsx")

def display_client_info(client_data):
    """Displays client information based on the given client data."""
    print(f"Client Name: {client_data['First Name']} {client_data['Last Name']}")
    print(f"Contact Number: {client_data['Contact Number']}")
    print(f"Wedding Date: {client_data['Wedding Date']}")
    print(f"Wedding Venue: {client_data['Wedding Vanue']}")
    print(f"Function Number of Days: {client_data['Function Number of days']}")
    print(f"Functions: {client_data['Functions']}")
    print(f"Photography Services: {client_data['Photography Services']}")
    print(f"Special Request: {client_data['Any Special Request or notes (optional)']}")

def find_client_by_phone(phone_number):
    """Finds a client by their phone number."""
    client_data = data[data['Contact Number'] == phone_number]
    if client_data.empty:
        print("Client not found.")
        return None
    else:
        return client_data.iloc[0]  # Assuming unique phone numbers

def generate_message(client_data, prices):
    first_name = client_data['First Name']
    last_name = client_data['Last Name']
    wedding_date = client_data['Wedding Date']
    venue = client_data['Wedding Vanue']
    days = client_data['Function Number of days']
    services = client_data['Photography Services'].split(', ')  # Split services into a list

    # Calculate the final price based on number of function days
    final_price = prices['total_price'] * days

    # Generate the message
    message = f"""
Cinematic Wedding Package

Dear {first_name} {last_name},

Your {days} days wedding photography package will be ₹{final_price}/- INR Only!

This is full wedding photography coverage for {days} days at {venue} with a team of 4-5 members.

Note -

1. Client has to take care of food and accommodation.
2. Our travel charges for your wedding are included in the package!
3. There will be 4-5 team members.

This includes -

{', '.join(services)}  # List services with commas

Deliverables -

1. All raw data will be provided. There's no photo limit for raw data. ️
2. Professional editing (colour correction and colour grading) on all selected pictures for Photobook. ️
3. Cinematic wedding highlight video of 4-5 minutes + 1 teaser of 30 sec - 1 minute. (Timeline depends on video data). ️️
4. Full HD traditional video of 2-3 hours. (Timeline depends on video data). ️️
5. 1 Premium luxury photobook of 45-50 (12×36 inch big size) sheets with leather printed cover and box.
6. 1 WhatsApp wedding invitation video. ️

Note - Client satisfaction is a must for us and we are very particular with our words!

T&C Apply

https://www.instagram.com/indoreweddingphotographer?igsh=Y25nbGR0NzE5a204
"""

    return message

def get_prices():
    prices = {}

    # Prompt the user for each service price with input validation
    while True:
        try:
            prices['candid_photography'] = int(input("Enter the price for Candid Photography: "))
            prices['cinematography'] = int(input("Enter the price for Cinematography: "))
            prices['traditional_photography'] = int(input("Enter the price for Traditional Photography: "))
            prices['traditional_videography'] = int(input("Enter the price for Traditional Videography: "))
            prices['drone_shooting'] = int(input("Enter the price for Drone Shooting: "))
            break  # Exit the loop if all inputs are valid
        except ValueError:
            print("Invalid input. Please enter numbers for prices.")

    # Calculate the total price
    prices['total_price'] = sum(prices.values())

    return prices

# Iterate over each client in the DataFrame
for index, row in data.iterrows():
    print(f"Processing client: {index}")  # Add this line for debugging
    client_data = row
    display_client_info(client_data)

    prices = get_prices()
    message = generate_message(client_data, prices)

    # Replace spaces with underscores in the filename
    file_name = f"{client_data['First Name'].replace(' ', '_')}_{client_data['Last Name'].replace(' ', '_')}_{client_data['Wedding Date'].strftime('%Y-%m-%d')}_wedding_package.txt"
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write(message)
```

```python
from googleapiclient.discovery import build
from google.oauth2.service_account import Credentials
import pandas as pd

# Define the scope
SCOPES = ['https://www.googleapis.com/auth/spreadsheets.readonly']

# Authenticate using the service account JSON key
creds = Credentials.from_service_account_file('path/to/scotlandyardproductions-credentials.json', scopes=SCOPES)

# Build the service
service = build('sheets', 'v4', credentials=creds)

# The ID and range of the spreadsheet
SAMPLE_SPREADSHEET_ID = 'your_spreadsheet_id_here'
SAMPLE_RANGE_NAME = 'Sheet1!A1:D10'  # Adjust range as needed

# Call the Sheets API
sheet = service.spreadsheets()
result = sheet.values().get(spreadsheetId=SAMPLE_SPREADSHEET_ID, range=SAMPLE_RANGE_NAME).execute()
values = result.get('values', [])

# Convert to DataFrame
df = pd.DataFrame(values)

# Print the data
print(df)
```
